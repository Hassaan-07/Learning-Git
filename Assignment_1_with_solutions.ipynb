```
 Write the pseudocode & the program to calculate GCD of two numbers.
```



````python
a,b = map(int,input("Enter both numbers : ").split())
c = b
rem = a%b
while rem: 
  if a<b:
    a = b
  b = rem
  rem = a%b


print(f"GCD of {a} and {c} => {b}" )
  

````

```
Write a program to count the number of digits in a number.
```



````python
n = int(input("Enter a number : "))
count=0
while n>9:
  n /= 10
  count+=1
else:
  print("No. of digits : " , count+1)
````



```
Write a program to reverse a number (9735 -> 5379)
```



````python
n = int(input("Enter a positive integer number : "))
rev = 0
while n>0:
  rm = n%10
  rev = rev*10 + rm
  n = n // 10
print("Mirror Image : ",rev)
````
